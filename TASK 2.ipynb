```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('indian_food.csv')
```

```python
print("Missing Values:\n", df.isnull().sum())
```

```python
df['state'] = df['state'].fillna('Unknown')  # Filling missing states as 'Unknown'
df['region'] = df['region'].fillna('Unknown')  # Filling missing regions as 'Unknown'
```

```python
df = df[(df['prep_time'] >= 0) & (df['cook_time'] >= 0)]
```

```python
df['flavor_profile'] = df['flavor_profile'].str.lower()
```

```python
print("\nDataset Information:\n")
df.info()
```

```python
print("\nSummary Statistics:\n", df.describe())
```

```python
# a) Categorical Data Analysis
# Bar Chart: Distribution of Diet Types
plt.figure(figsize=(8, 6))
df['diet'].value_counts().plot(kind='bar', color='purple')
plt.title('Distribution of Diet Types')
plt.xlabel('Diet')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Bar Chart: Distribution of Dishes by Course
plt.figure(figsize=(10, 6))
df['course'].value_counts().plot(kind='bar', color='lightblue')
plt.title('Distribution of Dishes by Course')
plt.xlabel('Course')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# b) Numerical Data Analysis
# Histograms of Prep Time and Cook Time
plt.figure(figsize=(10, 6))
plt.hist(df['prep_time'], bins=20, color='green', edgecolor='black')
plt.title('Distribution of Preparation Time')
plt.xlabel('Preparation Time (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['cook_time'], bins=20, color='orange', edgecolor='black')
plt.title('Distribution of Cooking Time')
plt.xlabel('Cooking Time (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# c) Boxplot to detect outliers in prep time and cook time
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['prep_time', 'cook_time']], palette='Set2')
plt.title('Boxplot of Preparation and Cooking Time')
plt.ylabel('Time (minutes)')
plt.show()
```

```python
# d) Correlation Analysis: Prep time vs Cook time
plt.figure(figsize=(10, 6))
sns.scatterplot(x='prep_time', y='cook_time', data=df, hue='diet', palette='Set1', alpha=0.7)
plt.title('Prep Time vs Cook Time (with Diet)')
plt.xlabel('Preparation Time (minutes)')
plt.ylabel('Cooking Time (minutes)')
plt.show()
```

```python
# e) Heatmap: Correlation between numerical features
plt.figure(figsize=(8, 6))
sns.heatmap(df[['prep_time', 'cook_time']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap between Prep Time and Cook Time')
plt.show()
```

```python
# f) Count of Dishes by Region
plt.figure(figsize=(10, 6))
df['region'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Number of Dishes by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
